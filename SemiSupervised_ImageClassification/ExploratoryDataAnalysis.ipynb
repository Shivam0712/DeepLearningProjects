## Exploratory Data Analysis

In this notebook we try to develop understanding about our data. In specific we check:
1. Train and Test data Size.
2. Distribution of Train Images across different categories.
3. We check images of flowers for different categories in Train Data.
4. We check images of flowers in Test Data

```python
# import useful packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from PIL import Image
import random
```

### Import Data

```python
# get path
path = os.getcwd()
```

```python
# import train csv
df_train = pd.read_csv(path+'/data/train.csv')

# import test csv
df_test = pd.read_csv(path+'/data/test.csv')

# check train data
df_train.head()
```

```python
# get stats
print('No. of samples in train data: ', len(df_train))
print('No. of samples in test data: ', len(df_test))
print('Total no. of categories in train data: ', len(df_train['category'].unique()))
```

```python
# distribution of train data cross categories
plt.hist(df_train['category'], bins = 102)
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Distribution of training data')
plt.savefig(path+'/data/plots/trainDistribution.png')
```

```python
# view some samples from train data
plt.figure(figsize=(20,10))
img_id = os.listdir(path+'/data/train/')
random.shuffle(img_id)
a = 0
for i in img_id[0:16]:
    a = a + 1
    img = Image.open(path+'/data/train/'+i)
    plt.subplot(4,4,a)
    plt.imshow(img)
plt.savefig(path+'/data/plots/trainSample.png')
```

```python
# view some samples from test data
plt.figure(figsize=(20,10))
img_id = os.listdir(path+'/data/test/')
random.shuffle(img_id)
a = 0
for i in img_id[0:16]:
    a = a + 1
    img = Image.open(path+'/data/test/'+i)
    plt.subplot(4,4,a)
    plt.imshow(img)
plt.savefig(path+'/data/plots/testSample.png')
```
